    Задание 1
    Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера.
    Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
movies = pd.read_csv('movies.csv')
movies
```

```python
top_rating = ratings[ratings['rating']==5].merge(movies, on='movieId', how='left')
```

```python
print('Больше всего оценок "5.0" У фильма: ', top_rating['title'].value_counts().head(1))
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года.
Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power
```

```python
pribalt_power = power[ ((power['country']=='Latvia') | (power['country']=='Lithuania') | (power['country']=='Estonia')) &\
                          ((power['category']==4) | (power['category']==12) | (power['category']==21)) &\
                          ((power['year']>=2005) & (power['year']<=2010)) & (power['quantity']>0) ]
print('Суммарное потребление стран Прибалтики категорий 4, 12 и 21 за период с 2005 по 2010 года составило: ',\
      pribalt_power['quantity'].sum())
```
